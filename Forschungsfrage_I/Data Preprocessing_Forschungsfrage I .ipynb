```python
#Create a new column for naming the labels
#In practise this would be the product category
#The product categories are = javascript = 1, java=2 , python=3, c#=4, others=5 
#We have decided that these are the products categroies which occure the most in the dataset; see Notebook "Data Exploration"
#The product category "others" can be reagarded as not product related questions from customers 

#Create a new column of NAN
sLength = len(selected_dataframe['title'])
selected_dataframe['product_category'] = pd.Series(np.nan, index=selected_dataframe.index)



#Create the labels_list (product categories)
produkt_liste=["(c#)","(python)" ,"(java)","(javascript)"]

#Search with Regex throuh the rows of the dataframe and label the product category
for i in range(len(selected_dataframe)):
    for x in range(len(produkt_liste)):
        if re.search(produkt_liste[x] ,selected_dataframe.loc[i,"tags"]):
            selected_dataframe.loc[i,"product_category"]=produkt_liste[x]



#transform the names into their product category number
for i in range(len(selected_dataframe)):
    if selected_dataframe.loc[i,"product_category"]=="(javascript)":
        selected_dataframe.loc[i,"product_category"]=0
        
    elif selected_dataframe.loc[i,"product_category"]=="(java)":
        selected_dataframe.loc[i,"product_category"]=1
        
    elif selected_dataframe.loc[i,"product_category"]=="(python)":
        selected_dataframe.loc[i,"product_category"]=2
        
    elif selected_dataframe.loc[i,"product_category"]=="(c#)":
        selected_dataframe.loc[i,"product_category"]=3
        
    else:
        selected_dataframe.loc[i,"product_category"]=4

#rename the column name from "product_category" to "product_category_number"
selected_dataframe.rename(columns={'product_category': 'product_category_number'})






selected_dataframe.head(50)
```

```python
#preprocess the dataframe for the test-train-split
selected_dataframe['sum']=selected_dataframe['title'].astype(str)+' '+selected_dataframe['body'].astype(str)+' '+selected_dataframe['tags'].astype(str)
preprocessed_dataframe= selected_dataframe.drop(['title','body','tags'],axis=1)
```
