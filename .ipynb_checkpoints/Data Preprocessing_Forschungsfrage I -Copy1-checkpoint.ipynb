```python
#Several Columns are not needed
#Dropped are: id_1, title_1, answer_count, score_1 ,body_1 , accepted_answer_id, score, comment_count, community_owned_date , creation_date, favorite_count, last_activity_date, last_edit_date ,last_editor_display_name, last_edit_date_1, last_editor_display_name_1, last_editor_user_id_1, owner_display_name_1, owner_user_id_1, parent_id_1, post_type_id_1, view_count, last_editor_user_id, owner_display_name, owner_user_id, parent_id, post_type_id, answer_count_1, answer_count_1, comment_count_1, community_owned_date_1, creation_date_1, favorite_count_1, last_activity_date_1, view_count_1, accepted_answer_id_1, tags_1
selected_dataframe = df.drop(['id_1','id','title_1','answer_count','score_1','body_1','accepted_answer_id', 'score','comment_count','community_owned_date','creation_date','favorite_count','last_activity_date','last_edit_date','last_editor_display_name','last_edit_date_1','last_editor_display_name_1','last_editor_user_id_1','owner_display_name_1','owner_user_id_1','parent_id_1','post_type_id_1','view_count','last_editor_user_id','owner_display_name','owner_user_id','parent_id','post_type_id','answer_count_1','answer_count_1','comment_count_1','community_owned_date_1','creation_date_1','favorite_count_1','last_activity_date_1','view_count_1','accepted_answer_id_1','tags_1'], axis=1)
```

```python
#Remaining data are: 
#The title = The headline of a Mail 
#The body = The Main content of a Mail
#The tags = The subjektline 

#Visualize the remaining info
selected_dataframe.info()
```

```python
#Visualize the remaining data head
selected_dataframe.head()
```

```python
#Create a new column for naming the labels
#In practise this would be the product category
#The product categories are = javascript = 1, java=2 , python=3, c#=4, php=5, android=6, c++=7, ios=8, others=9 
#We have decided that these are the products categroies which occure the most in the dataset; see Notebook "Data Exploration"
#The product category "others" can be reagarded as not product related questions from customers 

#Create a new column of NAN
sLength = len(selected_dataframe['title'])
selected_dataframe['product_category'] = pd.Series(np.nan, index=selected_dataframe.index)



#Create the labels_list (product categories)
produkt_liste=["ios+","c[+][+]","android","php","c#","python" ,"java","javascript"]

#Search with Regex throuh the rows of the dataframe and label the product category
for i in range(len(selected_dataframe)):
    for x in range(len(produkt_liste)):
        if re.search(produkt_liste[x] ,selected_dataframe.loc[i,"tags"]):
            selected_dataframe.loc[i,"product_category"]=produkt_liste[x]



#transform the names into their product category number
for i in range(len(selected_dataframe)):
    if selected_dataframe.loc[i,"product_category"]=="javascript":
        selected_dataframe.loc[i,"product_category"]=1
        
    elif selected_dataframe.loc[i,"product_category"]=="java":
        selected_dataframe.loc[i,"product_category"]=2
        
    elif selected_dataframe.loc[i,"product_category"]=="python":
        selected_dataframe.loc[i,"product_category"]=3
        
    elif selected_dataframe.loc[i,"product_category"]=="c#":
        selected_dataframe.loc[i,"product_category"]=4
        
    elif selected_dataframe.loc[i,"product_category"]=="php":
        selected_dataframe.loc[i,"product_category"]=5
        
    elif selected_dataframe.loc[i,"product_category"]=="android":
        selected_dataframe.loc[i,"product_category"]=6
        
    elif selected_dataframe.loc[i,"product_category"]=="c[+][+]":
        selected_dataframe.loc[i,"product_category"]=7
        
    elif selected_dataframe.loc[i,"product_category"]=="ios":
        selected_dataframe.loc[i,"product_category"]=8
    else:
        selected_dataframe.loc[i,"product_category"]=9

#rename the column name from "product_category" to "product_category_number"
selected_dataframe.rename(columns={'product_category': 'product_category_number'})







```
